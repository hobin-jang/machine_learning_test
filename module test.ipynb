```python
def And(x1,x2):
    w1,w2,theta = 0.5,0.5,0.7
    tmp = w1*x1+w2*x2
    if tmp<=theta:
        return False
    else:
        return True
```

```python
And(0,0)
And(0,1)
And(1,1)
```

```python
def step(x):
    return np.array(x > 0, dtype=np.int)
```

```python
import matplotlib.pyplot as plt
x=np.arange(-5.0,5.0,0.1)
y=step(x)
plt.plot(x,y)
plt.ylim(-0.1,1.1)
plt.xlim(-0.5,0.5)
```

```python
def sigmoid(x):
    return 1/(1+np.exp(-x))
```

```python
sigmoid(np.array([-1.0,1.0,5.0]))
```

```python
x=np.arange(-5,5,0.1)
y=sigmoid(x)
plt.plot(x,y)
```

```python
def relu(x):
    return np.maximum(0,x)
```
